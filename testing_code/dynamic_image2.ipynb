
# An animated image using a list of images


Animate an image from a list of images (or Artists).



```python
import matplotlib.pyplot as plt
import matplotlib.animation
import numpy as np

t = np.linspace(0,2*np.pi)
x = np.sin(t)

fig, ax = plt.subplots()
ax.axis([0,2*np.pi,-1,1])
l, = ax.plot([],[])

def animate(i):
    l.set_data(t[:i], x[:i])

ani = matplotlib.animation.FuncAnimation(fig, animate, frames=len(t))
```

```python
from IPython.display import HTML
HTML(ani.to_jshtml())
```

```python
import numpy as np
import matplotlib.pyplot as plt

fig, ax = plt.subplots()
ax.plot(np.random.rand(10))
text=ax.text(0, .05, "", va="bottom", ha="left")

def onclick(event):
    tx = 'button=%d, x=%d, y=%d, xdata=%f, ydata=%f' % (event.button, event.x, event.y, event.xdata, event.ydata)
    text.set_text(tx)

cid = fig.canvas.mpl_connect('button_press_event', onclick)
plt.show()
```

```python
def on_pick(event):
    artist = event.artist
    xmouse, ymouse = event.mouseevent.xdata, event.mouseevent.ydata
    x, y = artist.get_xdata(), artist.get_ydata()
    ind = event.ind
    print(f'Artist picked: {event.artist}')
    print(f'{len(ind)} vertices picked')
    print(f'Pick between vertices {min(ind)} and {max(ind)+1}')
    print(f'x, y of mouse: {xmouse:.2f},{ymouse:.2f}')
    print(f'Data point: {x[ind[0]]}, {y[ind[0]]}')
```

```python
fig, ax = plt.subplots()

tolerance = 10 # points
ax.plot(range(10), 'ro-', picker=tolerance)

fig.canvas.callbacks.connect('pick_event', on_pick)
plt.show()
```
