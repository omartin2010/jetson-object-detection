```python
import numpy as np
import cv2
import time
import os
import traceback
import matplotlib.pyplot as plt
from datetime import datetime, timedelta
```

```python
filename = os.path.join(os.getcwd(), 'output.avi')
print(filename)
```

```python
# cap = cv2.VideoCapture(1, cv2.CAP_V4L2)
# cap.set(3, 640)
# cap.set(4, 480)
```

```python
# cap = cv2.VideoCapture(1, cv2.CAP_V4L)
cap = cv2.VideoCapture(1, cv2.CAP_V4L)
# cap.set(3, 640)
# cap.set(4, 480)
try:
    fourcc = cv2.VideoWriter_fourcc(*'XVID')
    out = cv2.VideoWriter(filename, fourcc, 20.0, (1280, 720))
    start_time = datetime.now()
    print(f'Start time = {start_time.strftime("%Y-%m-%d %H:%M:%S.%f")}')
    while (datetime.now() - start_time).seconds < 2.254:
        ret, frame = cap.read()
        if ret==True:
            # write the flipped frame
            out.write(frame)
        else:
            print(f'Problem - ret-value = : {ret}')
            break
    if ret:
        end_time = datetime.now()
        print(f'Worked!')
        print(f'End time = {(end_time).strftime("%Y-%m-%d %H:%M:%S.%f")}')
        print(f'Effective Video time = {(end_time-start_time).seconds:.4f}s')
    # Release everything if job is finished
    cap.release()
    out.release()
except:
    print(f'Traceback : {traceback.print_exc()}')
```
